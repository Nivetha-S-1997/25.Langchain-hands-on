```python
#Loading libraries:
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.agents import Tool
from langchain_community.tools.tavily_search import TavilySearchResults

#Loading API KEYS:
load_dotenv()
gemini_api_key=os.getenv('GOOGLE_API_KEY')
tavily_api_key=os.getenv('TAVILY_API_KEY')

#Defining llm:
llm=ChatGoogleGenerativeAI(model='gemini-2.5-flash',temperature=0)

#Tool 1: QA tool:
qa_prompt=PromptTemplate.from_template("Answer clearly: {question}")
qa_chain=LLMChain(llm=llm,prompt=qa_prompt)

qa_tool=Tool(
    name='Simple QA',
    func=qa_chain.run,
    description='A basic LLM that answers user queries clearly'
)

#Tool 2: Summarizer tool:
summarize_prompt=PromptTemplate.from_template("Summarize the content: {content}")
summarize_chain=LLMChain(llm=llm,prompt=summarize_prompt)

summarizer_tool=Tool(
    name='Summarizer',
    func=summarize_chain.run,
    description='An LLM that summarizes the given content'
)

#Tool 3: Web search (Tavily):
search_tool=Tool(
    name='Tavily Search',
    func=TavilySearchResults(max_results=3).run,
    description='An LLM that searches in web to provide answer for users queries'
)

#Getting user queries:
qa_query='Near future of AI in 2 small paragraphs?'
summarize_content='''Dr. Reddy's Laboratories (or other major Pharmaceutical Companies like Cipla, Sun Pharma):
Why:The pharmaceutical industry is inherently a chemical industry. Companies like Dr. Reddys are investing heavily in AI 
for drug discovery, process optimization in manufacturing (e.g., optimizing reaction conditions, yield improvement), quality control, and 
supply chain management. Chemical engineers with a strong understanding of reaction kinetics, mass transfer, and process design are essential to 
integrate AI effectively into pharmaceutical R&D and production.'''
search_query='Latest updates on US-Iran war'

qa_response=qa_tool.run({'question':qa_query})
summarizer_response=summarizer_tool.run({'question':summarize_content})
search_response=search_tool.run({'question':search_query})

#Printing the results:
print("User query: ",qa_query)
print("\nLLM response: ",qa_response)
print("\nUser query: ",summarize_content)
print("\nLLM response: ",summarizer_response)
print("\nUser query: ",search_query)
print("\nLLM response: ",search_response)
```
